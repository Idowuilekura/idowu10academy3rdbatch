```python
from bs4 import BeautifulSoup
import requests 
page = requests.get("https://africafreak.com/100-most-influential-twitter-users-in-africa")
content = page.content
print(page.status_code)
```

```python
soup = BeautifulSoup(content,'html.parser')
print(soup)
```

```python
names = soup.find_all('h2')
```

```python
print(names)
```

```python
name = [i.get_text(strip=True) for i in names]
```

```python
print(len(name))
```

```python
print(name)
```

```python
name = name[-15:-5] # with hashtags
```

```python
name
```

```python
name = name[::-1] #reversing the names
```

```python
name # we need to extract the handles
```

```python
most_influential_handles =[]
uncleaned_handles = []
for i in name:
    splitted_name = i.split()
    index = len(splitted_name)-1
    uncleaned_handles.append(splitted_name[index])
for b in uncleaned_handles:
    remove_first = b.replace("(","") # remove first parenthesis
    remove_second = remove_first.replace(")","") # remove second parenthesis
    most_influential_handles.append(remove_second)

most_influential_handles

import pandas as pd
ten_most_influential_twitter_users_in_africa = pd.Series(most_influential_handles,index=range(1,11))
ten_most_influential_twitter_users_in_africa.to_csv("top_influentials_in_africa",encoding="utf-8",index=False)
```
